# Visualizing Two Quantative Variables

---

### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading Default Dataset

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

### Scatterplot

**Total bill vs. Tip**

```python
fig, ax = plt.subplots(figsize = (4, 2), dpi = 300)
sns.scatterplot(
    x = 'total_bill',
    y = 'tip',
    data = tips
)
```

**Number of male and females**

```python
fig, ax = plt.subplots(figsize = (4, 2),dpi = 300)

palette_col = {
    'Male' : 'tab:blue',
    'Female' : 'tab:red'
}

sns.countplot(
    y = "sex",
    data = tips,
    palette = palette_col
)

ax.set_xlabel('Number of observations', color = 'tab:purple')
ax.set_ylabel('Gender', color = 'tab:purple')
ax.set_title('Total Number of Male and Female patients', color = 'tab:purple')
plt.show()
```

### scatterplot 2

```python
sns.relplot(
    x = 'total_bill',
    y = 'tip',
    data = tips,
    kind = 'scatter',
    col = 'smoker'
)
```

```python
sns.relplot(
    x = 'total_bill',
    y = 'tip',
    data = tips,
    kind = 'scatter',
    row = 'smoker'
)
```

```python

sns.relplot(
    x = 'total_bill',
    y = 'tip',
    data = tips,
    kind = 'scatter',
    col = 'smoker',
    row = 'time'
)
```

```python
sns.relplot(
    x = 'total_bill',
    y = 'tip',
    data = tips,
    kind = 'scatter',
    col = 'day',
    col_wrap = 2
)
```

```python
sns.relplot(
    x = 'total_bill',
    y = 'tip',
    data = tips,
    kind = 'scatter',
    col = 'day',
    col_wrap = 2,
    col_order = ["Thur", "Fri", "Sat", "Sun"]
)
```
