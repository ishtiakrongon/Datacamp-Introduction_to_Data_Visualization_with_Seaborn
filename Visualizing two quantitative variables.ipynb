### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading Default Dataset

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

### Scatterplot

**Total bill vs. Tip**

```python
fig, ax = plt.subplots(figsize = (4, 2), dpi = 300)
sns.scatterplot(
    x = 'total_bill',
    y = 'tip',
    data = tips
)
```

**Number of male and females**

```python
fig, ax = plt.subplots()

palette_col = {
    'Male' : 'cyan',
    'Female' : 'tab:red'
}

sns.countplot(
    y = "sex",
    data = tips,
    palette = palette_col
)

ax.set_xlabel('Number of observations', color = 'tab:purple')
ax.set_ylabel('Gender', color = 'tab:purple')
ax.set_title('Total Number of Male and Female patients', color = 'tab:purple')
plt.show()
```

### scatterplot 2

```python
sns.relplot(
    x = 'total_bill',
    y = 'tip',
    data = tips,
    kind = 'scatter',
    col = 'smoker'
)
```

```python
sns.relplot(
    x = 'total_bill',
    y = 'tip',
    data = tips,
    kind = 'scatter',
    row = 'smoker'
)
```

```python

sns.relplot(
    x = 'total_bill',
    y = 'tip',
    data = tips,
    kind = 'scatter',
    col = 'smoker',
    row = 'time'
)
```

```python
sns.relplot(
    x = 'total_bill',
    y = 'tip',
    data = tips,
    kind = 'scatter',
    col = 'day',
    col_wrap = 2
)
```

```python
sns.relplot(
    x = 'total_bill',
    y = 'tip',
    data = tips,
    kind = 'scatter',
    col = 'day',
    col_wrap = 2,
    col_order = ["Thur", "Fri", "Sat", "Sun"]
)
```

### Loading another dataset

```python
student_data.head()
```

```python
# Change to use relplot() instead of scatterplot()
sns.relplot(x="absences", y="G3", 
                data=student_data, kind = 'scatter')

# Show plot
plt.show()
```

```python
# Change to make subplots based on study time
sns.relplot(x="absences", y="G3", 
            data=student_data,
            kind="scatter", col = "study_time")

# Show plot
plt.show()
```

```python
# Create a scatter plot of G1 vs. G3

sns.relplot(
    x = "G1",
    y = "G3",
    data = student_data,
    kind = "scatter"
)


# Show plot
plt.show()
```

```python
# Adjust to add subplots based on school support
sns.relplot(x="G1", y="G3", 
            data=student_data,
            kind="scatter", col = "schoolsup", col_order = ["yes", "no"])

# Show plot
plt.show()
```

```python
# Adjust further to add subplots based on family support
sns.relplot(x="G1", y="G3", 
            data=student_data,
            kind="scatter", 
            col="schoolsup",
            col_order=["yes", "no"],
            row = "famsup",
            row_order = ["yes", "no"])

# Show plot
plt.show()
```

### Drawing a relplot using size

```python
sns.relplot(
    x = 'total_bill',
    y = 'tip',
    data = tips,
    kind = 'scatter',
    size = 'size',
    hue = 'size'
)
```

### Drawing a relplot using style

```python
sns.relplot(
    x = 'total_bill',
    y = 'tip',
    data = tips,
    kind = 'scatter',
    style = 'sex',
    hue = 'sex'
)
```

### Drawing a relplot using alpha

```python
# Set alpha to be between 0 to 1
sns.relplot(
    y = 'total_bill',
    x = 'tip',
    data = tips,
    kind = 'scatter',
    alpha = 0.4
)
```

### Create a scatterplot horsepower vs. mpg

```python
sns.relplot(
    x = 'horsepower',
    y = 'mpg',
    data = mpg,
    kind = 'scatter',
    size = 'cylinders'
)
```

### Create a scatter plot of acceleration vs. mpg. Vary the style and color of the plot points by country of origin.

```python
sns.relplot(
    x = "acceleration",
    y = "mpg",
    data = mpg,
    kind = "scatter",
    style = "origin",
    hue = "origin"
)
```

### Interpreting Line Plots

#### Create a Line plot with model year on x-axis and mpg on y-axis.

```python
sns.relplot(
    x = 'model_year',
    y = 'mpg',
    data = mpg,
    kind = "line",
    ci = 'sd'
)
```

```python
# Create line plot of model year vs. horsepower
sns.relplot(
    x = "model_year",
    y = "horsepower",
    data = mpg,
    kind = 'line',
    ci = None
)
```

```python
# Change to create subgroups for country of origin
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None, style = "origin", hue = "origin", markers = True, dashes = False)

# Show plot
plt.show()
```
