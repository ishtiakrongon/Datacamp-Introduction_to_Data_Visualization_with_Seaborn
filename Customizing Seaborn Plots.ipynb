# Customizing Seaborn Plots

---

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## Loading Dataset

```python
country_data = pd.read_csv("D:/git_repositories/Datacamp-Introduction_to_Data_Visualization_with_Seaborn/Dataset/countries-of-the-world.csv", decimal = ',')
mpg = pd.read_csv("D:/git_repositories/Datacamp-Introduction_to_Data_Visualization_with_Seaborn/Dataset/mpg.csv")
student_data = pd.read_csv("D:/git_repositories/Datacamp-Introduction_to_Data_Visualization_with_Seaborn/Dataset/student-alcohol-consumption.csv", index_col = 0)
survey_data = pd.read_csv("D:/git_repositories/Datacamp-Introduction_to_Data_Visualization_with_Seaborn/Dataset/young-people-survey-responses.csv", index_col = 0)
tips = sns.load_dataset("tips")
```

---

### Changing Style and Palette

```python
tips.head()
```

```python
# Set the style to "whitegrid"

sns.set_style("whitegrid")
sns.set_palette("YlGnBu")

sns.catplot(x="day", 
            data= tips, 
            kind="count")

# Show plot
plt.show()
```

```python
# Set the context to "paper"

sns.set_context("paper")
sns.set_palette("PuRd")

# Create Bar Plot
sns.catplot(
    data = student_data,
    x = "study_time",
    y = "G3",
    kind = "bar"
)



# Show plot
plt.show()
```

```python
# Set the context to "notebook"

sns.set_context("notebook")
sns.set_palette("BuPu")

# Create Bar Plot
sns.catplot(
    data = student_data,
    x = "study_time",
    y = "G3",
    kind = "bar"
)



# Show plot
plt.show()
```

```python
# Set the context to "talk"
sns.set_context("talk")
sns.set_palette("Greens")
sns.set_style("darkgrid")

# Create Bar Plot
sns.catplot(
    data = student_data,
    x = "study_time",
    y = "G3",
    kind = "bar"
)

# Show plot
plt.show()
```

```python
# Set the context to "poster"
sns.set_context("poster")
sns.set_palette("Purples")
sns.set_style("dark")

# Create Bar Plot
sns.catplot(
    data = student_data,
    x = "study_time",
    y = "G3",
    kind = "bar"
)

# Show plot
plt.show()
```

```python
# Set the context

sns.set_context("notebook")

# Set the style to "darkgrid"

sns.set_style("darkgrid")

# Set a custom color palette

custom_palette = ["#39A7D0", "#36ADA4"]

sns.set_palette(custom_palette)

# Create the box plot of age distribution by gender
sns.catplot(x="Gender", y="Age", 
            data=survey_data, kind="box")

# Show plot
plt.show()
```
